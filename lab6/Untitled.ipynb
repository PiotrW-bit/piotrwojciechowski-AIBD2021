```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Original Data/2_LUBELSKIE.csv", index_col='Unnamed: 0')
df.info()
df.columns = ["Dni od zakupu", "Marka", "Wiek", "Płeć", "Ocena"]
df["Płeć"] = df['Płeć'].map({'bd.':None, 'K':'K', 'M':'M'})
```

```python
kobiety = len(df.loc[df['Płeć'] == 'K'])
faceci = len(df.loc[df['Płeć'] == 'M'])
nd = len(df.loc[df['Płeć'] == None])
Samsung = df.loc[(df.Marka == 'Samsung')]
fig, axs = plt.subplots(2,2, figsize=(15,10))
axs[0][0].hist(df["Ocena"], 11)
axs[0][0].set_title("Ocena")
axs[0][1].hist(df["Wiek"], 20)
axs[0][1].set_title("Wiek")
axs[1][0].bar(['K', 'M', 'nd.'], [kobiety, faceci, nd])
axs[1][0].set_title("Płeć")
```
