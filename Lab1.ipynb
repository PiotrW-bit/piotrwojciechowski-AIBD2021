Celem ćwiczenia było zdefiniowanie funkcji oraz z użyciem pakietu matplotlib zobrazowanie tej funkcji w odpowiednich zakresach. W drugiej części ćwiczenia zapoznawaliśmy się z biblioteką Pandas oraz klasą DataFrame, którą definiuje ta biblioteka.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def fun(x):
    return x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

y1 = fun(x1)
y2 = fun(x2)
y3 = fun(x3)

fig, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].plot(x1, y1)
axs[0].set_title("Wykres funkcji w zakresie (-1, 1)")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].legend(["x^2+5"])

axs[1].plot(x2, y2)
axs[1].set_title("Wykres funkcji w zakresie (-6, 6)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend(["x^2+5"])

axs[2].plot(x3, y3)
axs[2].set_title("Wykres funkcji w zakresie (-6, 6)")
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")
axs[2].legend(["x^2+5"])
plt.show()
```

Przy użyciu biblioteki numpy oraz matplotlib zobrazowałem wykres oraz opisałem wszystkie osie i legendy. W przypadku tych wykresów legenda nie była tak bardzo potrzebna, ponieważ w każdym  oknie wykresu znajduje się tylko jeden wykres. Nazwę funkcji można było wpisać jako tytuł wykresu lub tytuł subplotu.
